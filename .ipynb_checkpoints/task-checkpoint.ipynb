```python
#remove all duplicates and check shape
modified_df = df.drop_duplicates(keep = 'first')
modified_df.shape
```

```python
#Changing the mapping of columns 
modified_df = modified_df[(modified_df['status'] == 'WON') | (modified_df['status'] == 'LOST')]
modified_df['status'].value_counts()
```

```python
modified_df['status'] = modified_df['status'].map({'WON':1, 'LOST':0})
modified_df['status'].value_counts()
```

```python
# as task repplace'9b2d5b4678781e53038e91ea5324530a03f27dc1d0e5f6c9bc9d493a23be9de0' with NaN
modified_df.replace('9b2d5b4678781e53038e91ea5324530a03f27dc1d0e5f6c9bc9d493a23be9de0', np.NAN, inplace=True)
modified_df.head(3)
```

```python
#this fuction use to remove the columns having more than 40% missing values and define
def rmissingvaluecol(dff, threshold):
    col = []
    col = list(dff.drop(dff.loc[:,list((100*(dff.isnull().sum()/len(dff.index)) >= threshold))].columns, 1).columns.values)
    print("Columns having more than %s percent missing values: "%threshold, (dff.shape[1] - len(col)))
    print("Columns to be dropped                             : ", list(set(list((dff.columns.values))) - set(col)))
    return col

# Removing columns having 40% missing values
col = rmissingvaluecol(modified_df, 40)
modified_df = modified_df[col]
modified_df.head()
```

```python
#delete rows which contain either 70% or more than 70% Nan values
perc = 70.0 
min_count =  int(((100-perc)/100)*modified_df.shape[1] + 1)
modified_df = modified_df.dropna(axis=0, thresh=min_count)
modified_df.shape
```

```python
#checking for unique categoriesz
column_names = modified_df.columns
for column_name in column_names:
    print("Column Name        :", column_name)
    print("------------------------------------------")
    print(modified_df[column_name].value_counts(normalize=True, dropna=False)*100)
    print('\n')
```

```python
#remove columns which highly skewed data
skewed_columns_to_be_dropped = ['des_country','lead_id']

modified_df.drop(skewed_columns_to_be_dropped, axis=1, inplace=True)
modified_df.shape
```

```python
#calculate % of missing values
round(modified_df.isnull().sum() * 100 / len(modified_df), 2)
```

```python
modified_df.columns
```

```python
categorical_column_names = ['lost_reason', 'budget', 'lease', 'movein','source', 'source_city', 'source_country', 'des_city','utm_source','utm_medium']
for column_name in categorical_column_names:
    print("Column Name        :", column_name)
    print("------------------------------")
    #print("Unique Values      : ", modified_df[column_name].unique())
    
    values_to_be_imputed = modified_df[column_name].isnull().sum()
    print("Any Null (Before)  :", values_to_be_imputed)
    
    modified_df[column_name].fillna(modified_df[column_name].mode()[0], inplace=True)
    print(values_to_be_imputed, " values imputed with mode values of the column.")
    
    print("Null Values (After):", modified_df[column_name].isnull().sum())
    print('\n')
```

```python
column_names = modified_df.columns

for column_name in column_names:
    print("Column Name        :", column_name)
    print("-----------------------------------------------------")
    print(modified_df[column_name].value_counts(normalize=True, dropna=False)*100)
    print('\n')
```

```python
#Checking categories with lesser percentage to others
def change_to_others(x, value_counts_df):
    for key, val in value_counts_df.to_dict().items():
        if key == x and val < 10:
            return 'others'
    return x

# Checking the unique categories
columns_not_to_be_considered = ['status']
column_names = modified_df.columns
column_names = column_names.drop(columns_not_to_be_considered)
columns_to_be_changed = column_names

for column_name in columns_to_be_changed:
    print("Column Name : ", column_name)
    print("-----------------------------------------")

    value_counts_df = modified_df[column_name].value_counts(normalize=True) * 100
    print("Before :")
    print(value_counts_df)
    print('\n')

    modified_df[column_name] = modified_df[column_name].apply(lambda x:change_to_others(x, value_counts_df))
    value_counts_df = modified_df[column_name].value_counts(normalize=True) * 100
    print("After :")
    print(value_counts_df)
    print('\n')
```

```python
modified_df = modified_df.dropna()
```

```python
print("Before Cleaning Data")
print("*********************************")
print("Shape: ", df.shape)
print("Missing:")
print("-----------------------")
print(round(df.isnull().sum()*100/len(df), 2))

print('\n')
print("After Cleaning Data")
print("*********************************")
print("Shape: ", modified_df.shape)
print("Missing:")
print("-----------------------")
print(round(modified_df.isnull().sum()*100/len(modified_df), 2))
```

```python
column_names = ['Agent_id','lost_reason','budget','lease','movein','source','source_city','source_country','utm_source','utm_medium','des_city']

for column_name in column_names:
    dummies = pd.get_dummies(modified_df[column_name])
    dummies.drop('others', axis=1, inplace=True)
    modified_df = pd.concat([modified_df, dummies], axis=1)
    modified_df.drop(column_name, axis=1, inplace=True)
    print("Dummies created for: ", column_name)

modified_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
# Importing libraries for model preparing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import RandomForestClassifier

# Importing library for model building
import statsmodels.api as sm
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import RFE
from sklearn import metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7, test_size=0.3, random_state=100)
```

```python
plt.figure(figsize = (20,10))        
sns.heatmap(modified_df.corr(), annot = True, cmap="Reds")
plt.show()
```

```python
logm = sm.GLM(y_train,(sm.add_constant(X_train)), family = sm.families.Binomial())
logm.fit().summary()
```

```python
# Initializing LogisticRegression
logreg = LogisticRegression()

# Running RFE with 17 variables as output
rfe = RFE(logreg, step=17)             
rfe = rfe.fit(X_train, y_train)

# Listing the columns
for i in list(zip(X_train.columns, rfe.support_, rfe.ranking_)):
    print(i)
```

```python
# Listing down the columns important for building a model
col = X_train.columns[rfe.support_]
col
```

```python
# Listing down the columns not important
z = X_train.columns[~rfe.support_]
z
```

```python
X_train_sm = sm.add_constant(X_train[col])
logm1 = sm.GLM(y_train,X_train_sm, family = sm.families.Binomial())
res = logm1.fit()
res.summary()
```

```python
y_train_pred = res.predict(X_train_sm).values.reshape(-1)
y_train_pred[:10]
```

```python
y_train_pred_final = pd.DataFrame({'Converted':y_train.values, 'Converted_Prob':y_train_pred})
y_train_pred_final['LeadId'] = y_train.index
y_train_pred_final.head()
```

```python
y_train_pred_final['Predicted'] = y_train_pred_final.Converted_Prob.map(lambda x: 1 if x > 0.5 else 0)
# Let's see the head
y_train_pred_final.head()
```

```python
y_train_pred_final['Lead_Score'] = y_train_pred_final.Converted_Prob.map(lambda x: round(x*100))
y_train_pred_final.head()
```

```python
y_train_pred_final['Lead_Score'].value_counts()
```

```python
confusion = metrics.confusion_matrix(y_train_pred_final.Converted, y_train_pred_final.Predicted)
print(confusion)
```

```python
print(metrics.classification_report(y_train_pred_final.Converted, y_train_pred_final.Predicted))
```
